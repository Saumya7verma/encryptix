```python
1# prompt: write the code for creating AI tic tac toe program

import numpy as np

def create_board():
  """
  Creates a 3x3 board initialized with empty spaces.
  """
  return np.zeros((3, 3))

def is_valid_move(board, row, col):
  """
  Checks if a move is valid.
  """
  return board[row][col] == 0

def make_move(board, player, row, col):
  """
  Makes a move on the board for the given player.
  """
  board[row][col] = player

def check_win(board, player):
  """
  Checks if the given player has won the game.
  """
  win_conditions = [
    [board[0][0], board[0][1], board[0][2]],
    [board[1][0], board[1][1], board[1][2]],
    [board[2][0], board[2][1], board[2][2]],
    [board[0][0], board[1][0], board[2][0]],
    [board[0][1], board[1][1], board[2][1]],
    [board[0][2], board[1][2], board[2][2]],
    [board[0][0], board[1][1], board[2][2]],
    [board[2][0], board[1][1], board[0][2]],
  ]
  for condition in win_conditions:
    if all(cell == player for cell in condition):
      return True
  return False

def print_board(board):
  """
  Prints the current state of the board.
  """
  for row in board:
    for cell in row:
      if cell == 0:
        print(" ", end=" ")
      elif cell == 1:
        print("X", end=" ")
      elif cell == -1:
        print("O", end=" ")
    print()

def main():
  """
  Main function for the Tic-Tac-Toe game.
  """
  board = create_board()
  player = 1  # 1 for X, -1 for O
  while True:
    print_board(board)
    row = int(input("Enter row (0-2): "))
    col = int(input("Enter column (0-2): "))
    if not is_valid_move(board, row, col):
      print("Invalid move. Try again.")
      continue
    make_move(board, player, row, col)
    if check_win(board, player):
      print_board(board)
      print(f"Player {player} wins!")
      break
    player *= -1

if __name__ == "__main__":
  main()
```
