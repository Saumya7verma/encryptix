Load FLAN-T5 Model

```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

model_name = "google/flan-t5-large"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name).to(device)
```

Define Chat Function

```python

def chat_with_flan(prompt, max_length=512, temperature=0.7, repetition_penalty=1.2):
    inputs = tokenizer(prompt, return_tensors="pt").to(device)
    outputs = model.generate(
        **inputs,
        max_length=max_length,
        do_sample=True,           # enables randomness
        temperature=temperature,  # creativity
        repetition_penalty=repetition_penalty,
        top_p=0.9                 # nucleus sampling for more coherent output
    )
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response
```

Add Simple Chat Loop

```python
memory = []  # for basic short-term memory

while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit"]:
        break

    # Add previous conversation to prompt (memory)
    context = " ".join(memory[-5:])  # last 5 messages
    prompt = f"{context} \nUser: {user_input}\nAI:"

    answer = chat_with_flan(prompt)
    print("AI:", answer)

    # Save to memory
    memory.append(f"User: {user_input}")
    memory.append(f"AI: {answer}")
```
